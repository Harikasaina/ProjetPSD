### Enjeux de la personnalisation marketing
#### SWOT pour la Personnalisation Marketing
Forces :\
Données Abondantes : Accès à de grandes quantités de données clients pour une personnalisation précise.\
Technologies Avancées : Utilisation de l'IA pour analyser les données et automatiser les processus marketing.\
Personnalisation : Capacité à offrir des expériences client hautement personnalisées, augmentant l'engagement et la fidélité.\
Faiblesses :\
Confidentialité des Données : Risques liés à la collecte et à l'utilisation des données personnelles des clients.\
Dépendance Technologique : Besoin constant de technologies avancées et de compétences spécifiques pour les exploiter.\
Coûts Élevés : Investissements importants nécessaires pour mettre en place et maintenir des stratégies de personnalisation efficaces.\
Opportunités :\
Marchés Ciblés : Capacité à atteindre des segments de marché très spécifiques avec des messages personnalisés.\
Amélioration Continue : Utilisation des données pour continuellement améliorer et adapter les stratégies marketing.
Fidélisation Client : Augmentation de la fidélité client grâce à des expériences personnalisées.\
Menaces :\
Réglementations : Évolution des lois sur la protection des données pouvant impacter les stratégies de personnalisation.\
Concurrence : Risque que les concurrents adoptent des stratégies similaires ou plus efficaces.\
Saturation : Risque de saturation des clients avec trop de messages personnalisés, pouvant mener à une lassitude.\
#### 5P Marketing pour la Personnalisation
Produit : Adapter les offres produits en fonction des préférences et des comportements des clients.\
Prix : Proposer des prix personnalisés ou des promotions basées sur l'historique d'achat et les préférences.\
Place (Distribution) : Utiliser des canaux de distribution adaptés aux préférences des clients (en ligne, en magasin, etc.).\
Promotion : Créer des campagnes marketing personnalisées en utilisant les données clients pour des messages ciblés.\
Publicité : Utiliser l'IA pour cibler les publicités en ligne en fonction des comportements et des intérêts des clients.\
#### Parcours Client et Personnalisation
Découverte : Utiliser les données pour proposer des contenus et des publicités qui attirent l'attention des clients potentiels.\
Considération : Fournir des informations personnalisées pour aider les clients à évaluer les produits.\
Achat : Faciliter un processus d'achat fluide et personnalisé.\
Fidélité : Mettre en place des programmes de fidélité et des communications post-achat personnalisés pour encourager les achats répétés.\
Plaidoyer : Encourager les clients satisfaits à partager leurs expériences avec d'autres.\
#### Stratégie Marketing Digital Personnalisée Assistée par l'IA
Analyse de Données : Utiliser l'IA pour analyser les données clients et identifier les comportements, les préférences et les besoins.\
Segmentation : Segmenter les clients en groupes cibles basés sur les insights obtenus.\
Personnalisation des Contenus : Créer des contenus marketing personnalisés pour chaque segment.\
Automatisation : Utiliser l'IA pour automatiser les tâches marketing répétitives et améliorer l'efficacité.\
Évaluation Continue : Évaluer continuellement les performances des campagnes et ajuster les stratégies en fonction des résultats.\

## M3 Segmentation Client

```python
# Fusion des deux données
df=pd.read_csv('customers_data - customers_data.csv')
df1=pd.read_csv('sales_data - sales_data.csv')
merged_data=pd.merge(df, df1, on='Customer_ID')
merged_data.head(11)
```

```python
# Calcul des dépenses total par client
merged_data['depenseTtl']=merged_data['Quantity']*merged_data['Sale_Price']
depenseTtl=merged_data.groupby('Customer_ID')['depenseTtl'].sum().reset_index()
depenseTtl.columns=['Customer_ID','depenseTtl']
depenseTtl.head()
```

```python
Nbr_produit_unique=merged_data.groupby('Customer_ID')['Product_ID'].nunique().reset_index()
Nbr_produit_unique.columns=['Customer_ID','num_product_unique']
Nbr_produit_unique.head(10)
```

```python
final_df=pd.merge(df,depenseTtl,on ='Customer_ID')

final_df.head()
```

```python
final_df=pd.merge(final_df,Nbr_produit_unique, on='Customer_ID')
final_df.head()
```

```python
features=final_df[['Age','Total_Spent','num_product_unique','Gender']]

features.head()
```

```python
gender_encoded=pd.get_dummies(features['Gender'])
gender_encoded.head()
```

```python
feature=features.drop('Gender',axis=1)
feature.head()
```

```python
features=pd.concat([feature,gender_encoded],axis=1)
features.head()
```

```python
# normakisation dz vatiables numériques
scaler=StandardScaler()
```

```python
scaled_features=scaler.fit_transform(features[['Age','Total_Spent','num_product_unique']])
scaled_features
```

```python
cluster_features=pd.DataFrame(scaled_features,columns=['age_scaled','total_spent_scaled','num_unique_products_scaled'])
cluster_features.head()
```

```python
# calcul de l'erreur SSE pour diérents nombres de clusters
import os
os.environ["OMP_NUM_THREADS"] = "1"

sse=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(cluster_features[['age_scaled','total_spent_scaled','num_unique_products_scaled']])
    sse.append(kmeans.inertia_)


plt.title('Méthode du coude pour déterminer le nombre optimal de clusters')
plt.xlabel('nombre de cluster')
plt.ylabel('somme des carrés des erreurs(SSE)')
plt.show()
```
